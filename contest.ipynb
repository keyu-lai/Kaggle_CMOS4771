```python
import pandas as pd
import numpy
from IPython.display import display
from sklearn.feature_extraction import DictVectorizer
from sklearn.cross_validation import train_test_split

def encode_onehot(df, cols):
    vec = DictVectorizer()
    
    vec_data = pd.DataFrame(vec.fit_transform(df[cols].to_dict(orient='records')).toarray())
    vec_data.columns = vec.get_feature_names()
    vec_data.index = df.index
    
    df = df.drop(cols, axis=1)
    df = df.join(vec_data)
    return df

train = pd.read_csv('data.csv')
label = train['label']

X_train, X_test, y_train, y_test = train_test_split(train, label)
drop_cols = []
onehot_encoder = DictVectorizer(sparse=False)

for i in train.columns:
    if type(train[i].values[0]) == type(""):
        drop_cols = drop_cols + [i]

display(X_train)
display(X_test)

X_train = encode_onehot(X_train, drop_cols)
X_test = encode_onehot(X_test, drop_cols)


```
